For this version, we switch the primality test to Crypto.Util.number.isPrime since it implements Miller-Rabin test. Futhermore, we set the false positive rate to 1e-30 which gives us 50 rounds of Miller-Rabin.

### Utility functions

```python
def gen_safe_prime(param):
    while True:
        p = random_prime(2^param-1, false, 2^(param-1))
        if ZZ((p-1)/2).is_prime():
            return p
        
def is_prime_openssl(n):
    cmd = f"openssl prime {n}"
    process = subprocess.Popen(cmd.split(), stdout=subprocess.PIPE)
    output, _ = process.communicate()
    
    if output.strip()[-8:] == b'is prime':
        return true
    else:
        return false

def hash_to_prime(msg, out_len, hash_digest = 256):
        """
        Bari´c and Pfitzmann hash to prime function. I.e. compute 2^k*h(x)+r where h is a collision resistant hash function
        out_len: represent the bit length of the prime
        """
        
        bits_to_use = ZZ(floor(out_len/2)) #halving since we compute 2^bits_to_use * h(msg) + r; where |h(.)| = bits_to_use
        
        counter = ZZ(floor(bits_to_use/hash_digest))+1 #add 1 because 0 means we need to hash at least once
        
        bound = 2^ZZ(bits_to_use)
        
        hash_msg = bytearray()
        for i in range(1, counter+1):
            hash_msg += BLAKE2b.new(data=bytes(str(msg)+str(i), 'ascii'), digest_bits=hash_digest).digest()
            #hash_msg += BLAKE2s.new(data=bytes(str(msg)+str(i), 'ascii'), digest_bits=hash_digest).digest()
            
        int_msg = ZZ(int(bin(int.from_bytes(hash_msg, 'big'))[2:bits_to_use+2],2))
        #print(int_msg.nbits())
        ans = bound * int_msg
        num_trial = 1
        for r in range(1, bound):
            temp = ans + r
            if isPrime(int(temp), false_positive_prob=1e-30): # using Miller-Rabin with 50 rounds   
                ans = temp
                return (ans, r, num_trial)
            num_trial +=1
        return (-1,-1, num_trial)
    
def hash_to_prime_with_rnd(msg, rnd, out_len, hash_digest = 256):
        """
        Bari´c and Pfitzmann hash to prime function. 
        We already have the randomness rnd that supposedly give us a prime
        """
        bits_to_use = ZZ(floor(out_len/2)) #halving since we compute 2^bits_to_use * h(msg) + r; where |h(.)| = bits_to_use
        
        counter = ZZ(floor(bits_to_use/hash_digest))+1 #add 1 because 0 means we need to hash at least once
        
        bound = 2^ZZ(bits_to_use)
        
        hash_msg = bytearray()
        for i in range(1, counter+1):
            hash_msg += BLAKE2b.new(data=bytes(str(msg)+str(i), 'ascii'), digest_bits=hash_digest).digest()
            #hash_msg += BLAKE2s.new(data=bytes(str(msg)+str(i), 'ascii'), digest_bits=hash_digest).digest()
            
        int_msg = ZZ(int(bin(int.from_bytes(hash_msg, 'big'))[2:bits_to_use+2],2))

        #print(int_msg.nbits())

        ans = bound * int_msg + rnd
        if isPrime(int(ans), false_positive_prob=1e-30):
            return (ans, rnd)
        else:
            return (-1,-1)
    
def hash_to_prime_generic(msg, out_len, hash_digest=256):
    """
    Generic hash to prime function that compute H(m,r) and checks if the output is prime
    out_len: represent the bit length of the prime
    """
    trials = 2^out_len # size of the search binary space
    counter = ceil(out_len/hash_digest)
    block_length = floor(out_len/8) #numbers of bytes we actually need for the message

    num_trial = 1
    for i in range(trials):
        hash_msg = bytearray()

        for j in range(1, counter+1):
            hash_msg += BLAKE2b.new(data=bytes(str(msg)+str(i)+str(j), 'ascii'), digest_bits=hash_digest).digest()
            #hash_msg += BLAKE2s.new(data=bytes(str(msg)+str(i)+str(j), 'ascii'), digest_bits=hash_digest).digest()
        int_msg = ZZ(int(bin(int.from_bytes(hash_msg, 'big'))[2:out_len+2],2)) #cuts bits_number out of the hash value

        if isPrime(int(int_msg), false_positive_prob=1e-30): # using Miller-Rabin with 50 rounds  
            return (int_msg, i, num_trial)
        num_trial +=1
    return (-1, -1, num_trial)
    
def hash_to_large_odd_num(msg, out_len, hash_digest=256):
        """
        out_len: represents the length of the large number, recommend to start with 1024, and go with multiples of 256 
        """
        hash_msg = bytearray()
        counter = floor(out_len/hash_digest) #numbers of block of hash_digest-bit for the hash
        for i in range(1, counter+1):
            hash_msg+= BLAKE2b.new(data=bytes(str(msg)+str(i), 'ascii'), digest_bits=hash_digest).digest()
            #hash_msg+= BLAKE2s.new(data=bytes(str(msg)+str(i), 'ascii'), digest_bits=hash_digest).digest()
          
        hash_msg_final = bin(int.from_bytes(hash_msg, 'big')) #convert to binary 
        int_msg = ZZ(int('1'+hash_msg_final[3:len(hash_msg_final)-1]+'1', 2)) # replace the first and last bit by 1 and convert
       
        return int_msg
    
def gen_wit_from_set(base, list_element, sk, modulus):
    l = len(list_element) #assuming that it is a multiple of 2 for simplicity
    if l == 1:
        return [base]
    else:
        prod_1 = 1
        prod_2 = 1
        for elt in list_element[0: int(l/2)]:
            prod_1 *= elt

        for elt in list_element[int(l/2):l]:
            prod_2 *= elt

        prod_1 = prod_1%sk
        prod_2 = prod_2%sk

        base_1 = base.powermod(prod_1, modulus)
        base_2 = base.powermod(prod_2, modulus)
        left_witnesses = gen_wit_from_set(base_1, list_element[int(l/2):l], sk, modulus) 
        right_witnesses = gen_wit_from_set(base_2, list_element[0: int(l/2)], sk, modulus)
        return left_witnesses + right_witnesses
    
def gen_wit_from_set_iter(base, list_element, sk, modulus):
    stack = [(base, list_element)]
    results = []

    while stack:
        base, list_element = stack.pop()

        l = len(list_element)
        if l == 1:
            results.append(base)
            continue

        prod_1 = 1
        prod_2 = 1
        half_len = l // 2

        for elt in list_element[:half_len]:
            prod_1 *= elt

        for elt in list_element[half_len:]:
            prod_2 *= elt

        prod_1 %= sk
        prod_2 %= sk

        base_1 = base.powermod(prod_1, modulus)
        base_2 = base.powermod(prod_2, modulus)

        # Push the new problem onto the stack
        stack.append((base_2, list_element[:half_len]))  # Right half
        stack.append((base_1, list_element[half_len:]))  # Left half

    return results
```

```python
class RSA_Acc:
    def __init__(self, prime_bit_length):
        """
        Generate the parameter for an rsa accumulators
        return n with bits(n)=prime_bit_length*2, sk_1 = (p,q) where both are safe primes and bit length sec_param/2
            also return acc, the initial value of accumulator, which is a rand QR_n
        """
        if prime_bit_length == 1024:
            p = P1_1024
            q = P2_1024
        elif prime_bit_length == 2048:
            p = P1_2048
            q = P2_2048
        elif prime_bit_length == 1536:
            p = P1_1536
            q = P2_1536
            
        elif prime_bit_length == 3840:
            p = P1_3840
            q = P2_3840
            
        elif prime_bit_length == 7680:
            p = P1_7680
            q = P2_7680
        else:
            p = gen_safe_prime(ZZ(prime_bit_length))
            q = p
            while q==p :
                q = gen_safe_prime(ZZ(prime_bit_length))
                
        self.n = p*q
        self.sk_1 = (p,q)
        self.sk_2 = (p-1)*(q-1)
        self.acc = ZZ.random_element(1,self.n, distribution='uniform')
        if gcd(self.acc, self.n) != 1:
            self.acc = ZZ.random_element(1,self.n, distribution='uniform')
        self.acc = self.acc.powermod(2, self.n)
        self.base = self.acc # this represent the base value, before addition, of the accumulator
        self.sec_param = prime_bit_length
        
    def accumulate_without_hash(self, msg):
        temp = self.acc
        self.acc = self.acc.powermod(msg, self.n)
        return (temp, self.acc)
        
    def accumulate_without_hash_with_sk(self, msg):
        inv_msg = inverse_mod(msg, self.sk_2)
        w = self.base.powermod(inv_large_msg, self.n)
        return w
```

```python
class RSAprime(RSA_Acc):   
    def accumulate(self, msg, out_len, hash_digest=256):
        prime_msg = hash_to_prime(msg, out_len, hash_digest)
        #prime_msg = hash_to_prime_generic(msg, out_len, is_openssl)

        if prime_msg[0]== -1 and prime_msg[1] == -1:
            print("Error: failed to hash to prime")
            return

        temp = self.acc
        self.acc = self.acc.powermod(prime_msg[0], self.n)
        return (temp, self.acc, prime_msg[1])
            
    def accumulate_with_sk(self, msg, out_len, hash_digest=256):
        """
        Instead of computing acc^h(x), compute w = acc^{h(x)^-1}
        """ 
        prime_msg = hash_to_prime(msg, out_len, hash_digest)
        #prime_msg = hash_to_prime_generic(msg, out_len, is_openssl)

        if prime_msg[0]== -1 and prime_msg[1] == -1:
            print("Error: failed to hash to prime")
            return
        inv_prime_msg = inverse_mod(prime_msg[0], self.sk_2)
        w = self.acc.powermod(inv_prime_msg, self.n)
        return (w, prime_msg[1])
    
    def delete_without_witness(self, msg, rnd, out_len, hash_digest = 256):
        """
        Hash to prime using the randomness and then compute the inverse
        """
        prime_msg = hash_to_prime_with_rnd(msg, rnd, out_len, hash_digest)
        if prime_msg[0]== -1 and prime_msg[1] == -1:
            print("Error: failed to hash to prime")
            return
        inv_prime_msg = inverse_mod(prime_msg[0], self.sk_2)
        temp = self.acc.powermod(inv_prime_msg, self.n)
        self.acc = temp
        
    def delete_with_witness(self, msg, rnd, out_len, hash_digest = 256):
        pass
    
    def mem_verify(self, msg, rnd, w_msg, acc, out_len, hash_digest = 256):
        prime_msg = hash_to_prime_with_rnd(msg, rnd, out_len, hash_digest)
        if prime_msg[0]== -1 and prime_msg[1] == -1:
            print("Error: failed to hash to prime")
            return
        return w_msg.powermod(prime_msg[0], self.n) == acc
        

class RSANoPrime(RSA_Acc):
    def accumulate(self, msg, out_len, hash_digest=256):
        large_msg = hash_to_large_odd_num(msg, out_len, hash_digest=hash_digest)
        
        temp = self.acc
        self.acc = self.acc.powermod(large_msg, self.n)
        return (temp, self.acc)
        
    def accumulate_with_sk(self, msg, out_len, hash_digest=256):
        """
        Instead of computing acc^h(x), compute w = acc^{h(x)^-1}
        """  
        large_msg = hash_to_large_odd_num(msg, out_len, hash_digest)
        inv_large_msg = inverse_mod(large_msg, self.sk_2)
        w = self.acc.powermod(inv_large_msg, self.n)
        return w
    
    def delete_without_witness(self, msg, out_len, hash_digest = 256):
        large_msg = hash_to_large_odd_num(msg, out_len, hash_digest)
        inv_large_msg = inverse_mod(large_msg, self.sk_2)
        temp = self.acc.powermod(inv_large_msg, self.n)
        self.acc = temp
        
        
    def delete_with_witness(self, msg, w_msg, out_len, hash_digest=256):
        pass
    
    def mem_verify(self, msg, w_msg, acc, tau, out_len, hash_digest = 256):
        """
        tau represents 2^\tau as mentioned in the paper
        """
        large_msg = hash_to_large_odd_num(msg, out_len, hash_digest)
        w,S = w_msg
        for s in S:
            if s> tau:
                return False
        x = large_msg/math.prod(S)
        return w.powermod(x, self.n) == acc
    
    def batch_accumulate_sk(self, list_msg, out_len, hash_digest=256):
        """
        Compute prod(H(m)) root of A and return witness for each
        """
        list_msg_hash = []
        for msg in list_msg:
            list_msg_hash.append(hash_to_large_odd_num(msg, out_len, hash_digest))     
                                 
        total_prod = 1
        for msg in list_msg_hash:
            total_prod *= msg
            
#         print("total_prod")
#         print(total_prod)
#         print("\n")
        
        total_root_exp = inverse_mod(total_prod, self.sk_2)
        total_root = self.acc.powermod(total_root_exp, self.n)
#         print("total_root")
#         print(total_root)
#         print("\n")
        
        return gen_wit_from_set_iter(total_root, list_msg_hash, self.sk_2, self.n)
```

#### Average cost of a single Miller-Rabin Test on H_prime for a given p_len number

```python
time_miller_rabin = []
avg_of_avg_time_miller_rabin = []

def hash_to_prime_measure(msg, out_len, hash_digest = 256):
        """
        Bari´c and Pfitzmann hash to prime function. I.e. compute 2^k*h(x)+r where h is a collision resistant hash function
        out_len: represent the bit length of the prime
        """
        
        bits_to_use = ZZ(floor(out_len/2)) #halving since we compute 2^bits_to_use * h(msg) + r; where |h(.)| = bits_to_use
        
        counter = ZZ(floor(bits_to_use/hash_digest))+1 #add 1 because 0 means we need to hash at least once
        
        bound = 2^ZZ(bits_to_use)
        
        #internal_time_miller_rabin = []
        internal_time_miller_rabin_composite = []
        internal_time_miller_rabin_prime = []
        
        hash_msg = bytearray()
        for i in range(1, counter+1):
            hash_msg += BLAKE2b.new(data=bytes(str(msg)+str(i), 'ascii'), digest_bits=hash_digest).digest()
            #hash_msg += BLAKE2s.new(data=bytes(str(msg)+str(i), 'ascii'), digest_bits=hash_digest).digest()
            
        int_msg = ZZ(int(bin(int.from_bytes(hash_msg, 'big'))[2:bits_to_use+2],2))
        #print(int_msg.nbits())
        ans = bound * int_msg
        num_trial = 1
        for r in range(1, bound):
            temp = ans + r
            start = timer()
            b = isPrime(int(temp), false_positive_prob=1e-30)
            end = timer()
            #internal_time_miller_rabin.append(end-start)
            internal_time_miller_rabin_composite.append(end-start)
            if b: # using Miller-Rabin with 50 rounds   
                ans = temp
                internal_time_miller_rabin_prime.append(end-start)
                internal_time_miller_rabin_composite.pop()
                return internal_time_miller_rabin_composite, internal_time_miller_rabin_prime
            num_trial +=1
        return internal_time_miller_rabin_composite, internal_time_miller_rabin_prime
    

composite_test_time = []
prime_test_time = []

for m in test_list:
    comp_time, prime_time = hash_to_prime_measure(m, p_len, hash_digest)
    composite_test_time.extend(comp_time)
    prime_test_time.extend(prime_time)

print("Avg time, in seconds, needed to test a composite number: "+str(np.mean(composite_test_time))+ " (\U0001F643)")
print("Avg time, in seconds, needed to test a prime number:" + str(np.mean(prime_test_time)) + "(\U0001F642)")

print("\n")

print("Note that Number_of_primality*(\U0001F643) + (\U0001F642) dominates the execution time of H_prime")
print("See Table 1 in our paper to compare your results with ours")
```

#### Benchmarking Wesolowski PoE time

```python
# action performed by wesolowski prover
# c is the challenge that was produced by the hash function
def prover_poe_prime(v, u, e, statement, modu, out_len, hash_digest=256):
    #print("Wesolowski Prime")
    
    comp_statement = BLAKE2b.new(data=bytes(statement, 'ascii'), digest_bits= hash_digest).digest()
    t_1 = timer()
    c = hash_to_prime(comp_statement, out_len, hash_digest)
    s = floor(e/c[0])
    pi = v.powermod(s, modu)
    t_2 = timer()
    
    return pi, c[0], c[1], t_2-t_1

def prover_poe_no_prime(v, u, e, statement, modu, out_len, hash_digest=256):
    #print("Wesolowski No Prime")
    
    comp_statement = BLAKE2b.new(data=bytes(statement, 'ascii'), digest_bits=hash_digest).digest()
    t_1 = timer()
    c = hash_to_large_odd_num(comp_statement, out_len, hash_digest)
    s = floor(e/c)
    pi = v.powermod(s, modu)
    t_2 = timer()
    
    return pi, c, 1, t_2-t_1

def verifier_prime(pi, statement, rnd, v, u, e, modu, out_len, hash_digest=256):
    comp_statement = BLAKE2b.new(data=bytes(statement, 'ascii'), digest_bits= hash_digest).digest()
    
    t_1 = timer()
    c = hash_to_prime_with_rnd(comp_statement, rnd, out_len, hash_digest)
    r = e%c[0]
    u_prime = (pi.powermod(c[0], modu)*v.powermod(r, modu))%modu
    b = u_prime == u
    t_2 = timer()
    
    return b, t_2 - t_1

def verifier_no_prime(pi, statement, v, u, e, modu, out_len, hash_digest=256):
    comp_statement = BLAKE2b.new(data=bytes(statement, 'ascii'), digest_bits= hash_digest).digest()
    
    t_1 = timer()
    c = hash_to_large_odd_num(comp_statement, out_len, hash_digest=256)
    r = e%c
    u_prime = (pi.powermod(c, modu)*v.powermod(r, modu))% modu
    b = u_prime == u
    t_2 = timer()
    
    return b, t_2 - t_1
```

```python
prime_prove_time = []
noprime_prove_time = []
prime_ver_time = []
noprime_ver_time = []

for i in range(100):
    pi_odd, c_odd, r, podd_time =  prover_poe_no_prime(v, u_list[i], e_list[i], statement_list[i], 
                                                       modu, odd_len, hash_digest)
    pi_prime, c_prime, rnd, pprime_time = prover_poe_prime(v, u_list[i], e_list[i], statement_list[i],
                                                           modu, p_len, hash_digest)
    
    
    b_prime, vprime_time = verifier_prime(pi_prime, statement_list[i], rnd, v, u_list[i], e_list[i], modu, 
                                          p_len, hash_digest)
    b_odd, vodd_time = verifier_no_prime(pi_odd, statement_list[i], v, u_list[i], e_list[i], modu, odd_len, hash_digest)
    
    prime_prove_time.append( pprime_time)
    noprime_prove_time.append(podd_time)
    prime_ver_time.append(vprime_time)
    noprime_ver_time.append(vodd_time)
    
    if (i+1)%10 == 0:
        print(i+1)
```
